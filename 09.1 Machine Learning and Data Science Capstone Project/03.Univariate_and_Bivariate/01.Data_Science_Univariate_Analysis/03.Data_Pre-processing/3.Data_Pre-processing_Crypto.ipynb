```python
""" 
For Regeression:
Existing Four Scenarios for this Dataset to Replace the NaN/Null values in Column-Salary/Any Column :
1.Delete Entire Row(s) those having NaN/Null values in Column-Salary
2.Create a Model by considering this Dataset as Semi-supervised Learning, so that We can Predict Values 
for NaN/Null values in Column-Salary
3.Replace the NaN/Null values in Column-Salary with Central Tendency by Taking Mean/Median/Mode
4.Replace the NaN/Null values in Column-Salary with respective to Problem Statement (Requirement/Need) 

For Categorical Data:
Existing Scenario for this Dataset to Replace the NaN/Null values in Column-Salary/Any Column :
1.Replace the NaN/Null values in Column-Salary with Central Tendency by Taking Mode """
```

```python
#Purpose : To Automate the Separation of Both Quantitive and Qualitative Columns with Return Statements of Created List(s) - quan, qual

def quanQual(dataset):
        quan = []
        qual = []

        for columnName in dataset.columns:
            if dataset[columnName].dtype == 'O':
                qual.append(columnName)
                  
            elif dataset[columnName].dtype == 'bool':  # Boolean type
                qual.append(columnName)
            
            elif 'datetime' in str(dataset[columnName].dtype):  # Datetime types
                qual.append(columnName)
                
            elif dataset[columnName].dtype in ['int64', 'float64']:  # Numeric types
                quan.append(columnName)
            
            else:
                qual.append(columnName)
        
        return quan, qual
```

```python
# Assigning the Output of Function - quanQual into Two New Variables - quan and qual, Since the Function returns Two Outputs

quan, qual = quanQual(dataset)
```

```python
""" How to Handle/Automate the Central Tendency (Taking Mean/Median/Mode) for Multiple Column
Solution: Using "Imputer in Python"
Purpose: Imputation of Missing Values """

#importing Library - numpy and assigning into a New Variable - np
import numpy as np 

#importing a Class - SimpleImputer from module - impute of Library - sklearn 
from sklearn.impute import SimpleImputer

#Calling the Method - SimpleImputer (With Parameters) assigning into a New Variable - imp
#Creating imputer which is configured to replace missing values (np.nan) with the mean of each column
imp = SimpleImputer(missing_values=np.nan, strategy="mean")

# This step calculates the mean of each column in 'quan' and prepares the imputer to replace missing values
imp.fit (dataset[quan])
imp
```

```python
# Transforming the data by replacing missing values with the computed means
imputed_data = imp.transform (dataset[quan])
imputed_data
```

```python
# Convert the NumPy array back to a pandas DataFrame
imputed_dataset = pd.DataFrame(imputed_data, columns=quan)
imputed_dataset
```

```python
#Method-01, To check the Presence of Null values in Entire Imputed_dataset in Count, since having poor readability in Python
imputed_dataset.isnull().sum()
```

```python
#Method-02, To check the Presence of Null values in Entire Imputed_dataset in Count, since having poor readability in Python
imputed_dataset.isna().sum()
```

```python
categorical_dataset = dataset[qual]
categorical_dataset
```

```python
#Assigning the Two Different Datasets into New List - Two_Datasets

Two_Datasets = [categorical_dataset,imputed_dataset]
```

```python
""" Concatinating the Two Different Datasets present in List - Two_Datasets with Parameter (axis=1) 
to set the merge function to be happened Column wise"""

Preprocessed_Dataset = pd.concat(Two_Datasets,axis=1)
Preprocessed_Dataset
```

```python
#Saving the Concatinated Dataset as .csv File

Preprocessed_Dataset.to_csv("Preprocessed_Data_Cryptos.csv",index=False)
```
